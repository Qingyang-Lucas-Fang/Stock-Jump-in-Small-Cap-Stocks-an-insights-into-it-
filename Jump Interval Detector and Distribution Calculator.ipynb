# Stats Distribution Fitting

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import r2_score

# --- STEP 1: DOWNLOAD DATA ---
 # Shortened list for example speed
raw_data = pd.DataFrame()
for name in sample_stocks:
    try:
        data = yf.download(name, interval='1h', start='2024-01-01', end='2025-01-01')
        if not data.empty:
            raw_data[name] = data['Close']
    except Exception as e:
        print(f"Skipping {name}: {e}")

# --- STEP 2: PREPARE DATA (WIDE TO LONG) ---
raw_data_long = raw_data.melt(ignore_index=False, var_name='Ticker', value_name='Close')
raw_data_long.index = pd.to_datetime(raw_data_long.index)

# --- STEP 3: ANALYZE JUMPS ---
def analyze_jump_intervals(df):
    # 1. Identify Jumps using Lee-Mykland logic
    df = df.sort_values(['Ticker', df.index.name if df.index.name else 'index'])
    df['Returns'] = df.groupby('Ticker')['Close'].transform(lambda x: np.log(x / x.shift(1)))

    # Robust Volatility (Bipower Variation)
    def get_local_vol(group):
        abs_ret = np.abs(group)
        # Prevents error if group is too small
        bv = (np.pi / 2) * (abs_ret * abs_ret.shift(1)).rolling(window=90, min_periods=10).mean()
        return np.sqrt(bv)

    df['Local_Vol'] = df.groupby('Ticker')['Returns'].transform(get_local_vol)
    df['L_stat'] = df['Returns'] / df['Local_Vol']
    df['Is_Jump'] = np.abs(df['L_stat']) > 4.0

    # 2. Extract Timestamps of Jumps
    # We reset index here so 'Date' becomes a column we can manipulate
    jump_times = df[df['Is_Jump'] == True].copy().reset_index()
    date_col = jump_times.columns[0] # Usually 'Datetime' or 'Date'

    # 3. Calculate Intervals per Ticker
    # Use groupby on the Ticker and shift the actual date column
    jump_times['Prev_Jump'] = jump_times.groupby('Ticker')[date_col].shift(1)
    jump_times['Interval_Hours'] = (jump_times[date_col] - jump_times['Prev_Jump']).dt.total_seconds() / 3600

    # Filter out NaNs and zero-intervals
    intervals = jump_times['Interval_Hours'].dropna()
    intervals = intervals[intervals > 0].values

    return intervals

# Run Analysis
ijt_data = analyze_jump_intervals(raw_data_long)

# --- STEP 4: DISTRIBUTION FITTING ---
if len(ijt_data) > 0:
    plt.figure(figsize=(12, 6), facecolor='white')
    counts, bin_edges, _ = plt.hist(ijt_data, bins=50, density=True, alpha=0.3, color='gray', label='Raw Intervals')
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

    # Fit 1: Exponential
    expon_params = stats.expon.fit(ijt_data)
    expon_pdf = stats.expon.pdf(bin_centers, *expon_params)

    # Fit 2: Weibull
    weibull_params = stats.weibull_min.fit(ijt_data)
    weibull_pdf = stats.weibull_min.pdf(bin_centers, *weibull_params)

    # Fit 3: Lognormal
    lognorm_params = stats.lognorm.fit(ijt_data)
    lognorm_pdf = stats.lognorm.pdf(bin_centers, *lognorm_params)

    # Plotting
    plt.plot(bin_centers, expon_pdf, 'r--', lw=2, label=f'Exponential R²={r2_score(counts, expon_pdf):.4f}')
    plt.plot(bin_centers, weibull_pdf, 'g-', lw=2, label=f'Weibull R²={r2_score(counts, weibull_pdf):.4f}')
    plt.plot(bin_centers, lognorm_pdf, 'b-', lw=2, label=f'Lognormal R²={r2_score(counts, lognorm_pdf):.4f}')

    plt.title("Distribution of Wait Times Between Jumps", fontsize=14)
    plt.xlabel("Wait Time (Hours)")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(alpha=0.2)
    plt.show()
else:
    print("No jumps detected with the current threshold.")
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

def calculate_metrics(data, dist_name, params):
    dist = getattr(stats, dist_name)
    # Use logpdf to calculate log-likelihood
    # We add a small epsilon to data to avoid log(0) errors if any exist
    log_likelihood = np.sum(dist.logpdf(data, *params))

    k = len(params)
    n = len(data)

    # Check for infinity/NaN in log_likelihood
    if np.isinf(log_likelihood) or np.isnan(log_likelihood):
        return -np.inf, np.inf, np.inf

    aic = 2 * k - 2 * log_likelihood
    bic = k * np.log(n) - 2 * log_likelihood

    return log_likelihood, aic, bic

# --- Improved Analysis Section ---
if len(ijt_data) > 0:
    # 1. CLEAN DATA: Remove any non-positive values that break Weibull/Lognormal
    clean_data = ijt_data[ijt_data > 0]

    plt.figure(figsize=(12, 7), facecolor='white')
    counts, bin_edges, _ = plt.hist(clean_data, bins=100, density=True, alpha=0.3, color='gray')
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

    # Fit Distributions with FIXED LOCATION (floc=0)
    # This prevents the 'Infinity' AIC error by forcing the distribution to start at 0
    results_table = []

    # Define fits: (Dist_Name, Label, Color, Style)
    fit_configs = [
        ('expon', 'Exponential', 'red', '--'),
        ('weibull_min', 'Weibull', 'green', '-'),
        ('lognorm', 'Lognormal', 'blue', '-')
    ]

    for dist_name, label, color, style in fit_configs:
        dist = getattr(stats, dist_name)

        # KEY ADJUSTMENT: Fix location to 0 (floc=0)
        if dist_name == 'expon':
            params = dist.fit(clean_data, floc=0)
        else:
            params = dist.fit(clean_data, floc=0)

        # Calculate Metrics
        ll, aic, bic = calculate_metrics(clean_data, dist_name, params)
        pdf = dist.pdf(bin_centers, *params)
        r2 = r2_score(counts, pdf)

        results_table.append({
            'Distribution': label,
            'AIC': aic,
            'BIC': bic,
            'Log-Likelihood': ll,
            'R2': r2,
            'Shape_Param': params[0] if len(params) > 2 else 'N/A'
        })

        plt.plot(bin_centers, pdf, color=color, linestyle=style, lw=2.5,
                 label=f"{label} (AIC: {round(aic, 1)})")

    plt.title("Jump Interval Analysis: Fixed-Location (floc=0) Comparison", fontsize=14)
    plt.xlabel("Wait Time (Hours)")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(alpha=0.15)
    plt.show()

    # Display Results
    results_df = pd.DataFrame(results_table).sort_values('AIC')
    print("\n--- Model Selection Criteria (Sorted by AIC) ---")
    print(results_df.to_string(index=False))
```
